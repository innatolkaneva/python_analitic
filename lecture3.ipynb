```python
age_days = []
for i, row in df.iterrows():
    age_days.append(row['Age'] *365)
age_days[:10]
```

```python
df['age_days2'] = age_days
df.head()
```

```python
def age_to_days(x):
    return x*365
df['age_days3'] = df['Age'].apply(age_to_days)
df.head()
```

```python
df = df.drop(columns = 'age_days3')
df.head()
```

```python
df.drop(columns = 'Age_days', inplace = True)
df.head()
```

```python
import time
from tqdm import tqdm
tqdm.pandas()
def age_to_pr(x):
    return x*365
    time.sleep(0.001)
df['Age'].progress_apply(age_to_pr)
```
