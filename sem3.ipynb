```python
df['Cpu_company'] = df['Cpu'].apply(lambda x : x.split()[0])
df.head()
```

```python
df['Cpu_company'] = df['Cpu'].str.split().str.get(0)
df.head()
```

```python
df['Memory'].unique()
```

```python
def split_memory(text):
    words = text.split()
    if words[0].endswith('TB'):
        result = int(float(words[0].replace('TB',''))*1024)
    else:
        result =  int(float(words[0].replace('GB','')))
    if '+' in words:
        if words[words.index('+')+1].endswith('TB'):
            result = result + int(float(words[words.index('+')+1].replace('TB',''))*1024)
        else:
            result = result + int(float(words[words.index('+')+1].replace('GB','')))
    return result
df['Summ_memory'] = df['Memory'].apply(split_memory)
df.head(34)
    
```

```python
df['Summ_memory'] = df['Memory'].apply(lambda i:int(i.split()[0].replace('GB','')) if 'GB' in i else int(float(i[0].replace('TB',''))*1024))
df.head()
```
